```python
import json
import requests
from bs4 import BeautifulSoup
import html2text

# Load JSON file
with open("guidelines_database.json", "r", encoding="utf-8") as f:
    data = json.load(f)

# Initialize Markdown converter
converter = html2text.HTML2Text()
converter.ignore_links = False  # Preserve links
converter.ignore_images = False  # Preserve images
converter.ignore_tables = False  # Preserve tables

# Function to scrape content
def scrape_content(url):
    try:
        headers = {"User-Agent": "Mozilla/5.0"}
        response = requests.get(url, headers=headers, timeout=10)
        response.raise_for_status()
        
        soup = BeautifulSoup(response.text, "html.parser")

        # Try to find the main content (may need to tweak based on website structure)
        main_content = soup.find("article") or soup.find("div", {"class": "content"}) or soup.body
        if not main_content:
            return None
        
        return converter.handle(str(main_content))  # Convert to Markdown
    except requests.exceptions.RequestException as e:
        print(f"Failed to fetch {url}: {e}")
        return None

# Process each entry
markdown_texts = []
for sub in data:
    url = sub["PubMed"] if sub["PubMed"] else sub["URL"]
    if url:
        print(f"Scraping: {url}")
        markdown_content = scrape_content(url)
        if markdown_content:
            markdown_texts.append(markdown_content)
    break

# Save to a Markdown file
if markdown_texts:
    with open("guidelines.md", "w", encoding="utf-8") as f:
        f.write("\n\n---\n\n".join(markdown_texts))  # Separate articles with a divider

print("Markdown file saved as guidelines.md")
```
